```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status': 'Shaadi'})
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

# Exploratory Data Analysis

Gender

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

colors = ['#FF6347', '#4682B4']  # custom colors

plt.bar(sales_gen['Gender'], sales_gen['Amount'], color=colors)
plt.show()
```

***From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.***

Age

```python
df.columns
```

```python
ax = sns.countplot(x='Age Group', data=df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

colors = ["#FF6F61", "#6B5B95", "#88B04B", "#F7CAC9", "#92A8D1", "#955251", "#B565A7", "#009B77", "#DD4124", "#45B8AC"]

plt.bar(sales_age['Age Group'], sales_age['Amount'], color=colors)
plt.show()
```

***From above graphs we can see that most of the buyers are of age group between 26-35 yrs female.***

State

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17, 5)})
sns.barplot(x='State', y='Orders', data=sales_state, hue='State')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17, 5)})
sns.barplot(x='State', y='Amount', data=sales_state, hue='State')
```

***From above graphs we can see that unexpectedly most of the orders are from Uttar Pradesh, Maharashtra and Karnataka.***

Marital Status

```python
ax = sns.countplot(x='Marital_Status',  data=df, hue='Marital_Status')

sns.set(rc={'figure.figsize':(7, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital_status = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(7, 5)})
sns.barplot(x='Marital_Status', y='Amount', data=sales_marital_status, hue='Gender')
```

***From above graphs we can see that most of the buyers are married (women) and they have high purchasing power.***

Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Occupation', data=df, hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data=sales_occupation, hue='Occupation')
```

***From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation Sector.***

Product Category

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(x='Product_Category', data=df, hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category', y='Amount', data=sales_product_category, hue='Product_Category')
```

***From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category.***

```python
df.columns
```

```python
sales_product_ID = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID', y='Orders', data=sales_product_ID, hue='Product_ID')
```

# Conclusion:

***Married women age group 26-35 yrs from Uttar Pradesh, Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.***
